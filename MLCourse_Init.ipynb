```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
veri = pd.read_excel("/content/drive/MyDrive/dataset_Aylik.xlsx")
```

```python
print(veri)
```

```python
veri.info()
veri.value_counts("Month")
```

```python
ortalama_veriler = veri.groupby("Month").mean()


ortalama_veriler = ortalama_veriler.sort_values("W (gr)", ascending=False)
print(ortalama_veriler)

# Eğer ay isimleri string olarak varsa:
aylar = ["Ocak", "Şubat", "Mart", "Nisan", "Mayıs", "Haziran", "Temmuz", "Agustos", "Eylül", "Ekim", "Kasım", "Aralık"]

df_grouped = veri.groupby("Month").mean()
df_grouped = df_grouped.reindex(aylar)
print(df_grouped)
```

```python
veri.count()
```

```python
print(veri["Month"].dtype)
print(veri["Month"].unique())

aylar = {
    "Ocak": 1, "Şubat": 2, "Mart": 3, "Nisan": 4, "Mayıs": 5, "Haziran": 6,
    "Temmuz": 7, "Agustos": 8, "Eylül": 9, "Ekim": 10, "Kasım": 11, "Aralık": 12
}

veri["Month"] = veri["Month"].map(aylar)
veri.mean()
```

```python
veri.std()
```

```python
print(veri.min())
print(veri.max())
```

```python
veri.describe()
```

```python
plt.plot(veri['LT (cm)'],veri['W (gr)'],'bx')
plt.xlabel('LT (cm)')
plt.ylabel('W (gr)')
plt.title('LT ve W Grafiği')

plt.show()
```

Scatter grafikleri dağılımları göstermesi açısından faydalı o sebeple datanın dağılımını anlamakta önemli olabilir.


```python
x = veri['LT (cm)']
y = veri['W (gr)']
plt.scatter(x,y)
plt.xlabel('LT (cm)')
plt.ylabel('W (gr)')
plt.title('LT ve W Grafiği')
plt.show()


x = veri['Month']
y = veri['W (gr)']
plt.scatter(x,y,color='g')
plt.xlabel('Month')
plt.ylabel('W (gr)')
plt.title('Month ve W Grafiği')
plt.show()
```

**3D Scatter plot**

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = veri['Month']
y = veri['LT (cm)']
z = veri['W (gr)']
ax.scatter(x, y, z, c='g', marker='x')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap='Greens')
plt.show()
```
